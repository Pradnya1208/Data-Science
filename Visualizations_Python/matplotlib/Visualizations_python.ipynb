# Matplotlib

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
x = [-3,5,7]
y = [10,2,5]
```

```python
fig = plt.figure(figsize =(15,3))
plt.plot(x,y)
plt.xlim(-4,10)
plt.ylim(0,12)
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Line Plot')
plt.suptitle('Sales Comparison', size =20, y = 1.05, x = 0.51)
```

```python
fig.get_size_inches()
```

```python
fig.set_size_inches(14,4)
fig
```

```python
import pandas as pd
import numpy as np

cars = pd.read_csv('mtcars.csv')
cars.head()
```

```python
cars.shape
```

```python
pd.crosstab(cars.gear,cars.cyl)
```

## Bar Plot

```python
pd.crosstab(cars.gear,cars.cyl).plot(kind="bar")


plt.xlabel('Gear')
plt.ylabel('Number of Gears')
plt.title('Plot for Gears and Cylinder details of cars')
```

```python
pd.crosstab(cars.gear,cars.carb)
```

```python
pd.crosstab(cars.gear,cars.carb).plot(kind="bar")

plt.xlabel('Gear')
plt.ylabel('Number of Gears')
plt.title('Plot for Gears and carburettor details of cars')
```

## Pie Chart 

```python
cars["gear"].value_counts()

cars.gear.value_counts().plot(kind="pie")
plt.title('Pie plot for No. of gears in cars')
```

```python
cars["carb"].value_counts()

cars.carb.value_counts().plot(kind="pie")
plt.title('Pie plot for carburettors')
```
