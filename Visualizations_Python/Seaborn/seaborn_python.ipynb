```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

### Univariate data visualization

```python
#strip plot
sns.stripplot(y ='tip', data = tips , jitter =True)
```

```python
sns.stripplot(x= 'day', y ='tip', data = tips , jitter =True)
```

```python
sns.stripplot(x= 'smoker', y ='tip', data = tips , jitter =True) # smoker is a categorical variable 
```

```python
sns.stripplot(x= 'sex', y ='tip', data = tips , jitter =True)
```

```python
sns.stripplot(x= 'sex', y ='tip', data = tips , jitter = False)
```

```python
# swarm plot
#for clear idea of density as compared to scatter plot
sns.swarmplot(x= 'day', y ='tip', data = tips)
```

```python
sns.swarmplot(x= 'day', y ='tip', data = tips, hue ='sex')
```

## Box and Violin plot

```python
sns.boxplot(x= 'day', y ='tip', data = tips)
```

```python
sns.violinplot(x= 'day', y ='tip', data = tips)
```

## Bivariate plots

```python
#joint plot
sns.jointplot(x= 'total_bill', y= 'tip', data =tips)
```

```python
#Density plot
sns.jointplot(x= 'total_bill', y= 'tip', data =tips, kind = 'kde')
```

```python
#pair plot
sns.pairplot(data = tips)
```

```python
sns.pairplot(data = tips, hue ='sex')
```
