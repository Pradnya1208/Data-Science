```python
x = [-3,5,7]
y = [10,2,5]
```

```python
fig = plt.figure(figsize =(15,3))
plt.plot(x,y)
plt.xlim(-4,10)
plt.ylim(0,12)
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Line Plot')
plt.suptitle('Sales Comparison', size =20, y = 1.05, x = 0.51)
```
